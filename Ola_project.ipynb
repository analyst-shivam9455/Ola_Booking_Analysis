# OLA_Booking_Analysis

```python
# importing necessary libraries
import pandas as pd
import datetime
```

```python
# loading data
df = pd.read_csv('Bookings-20000-Rows.csv')
df.head(2)
```

```python
# converting date column into datatime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# make two new column using date column
df['Booking_Date'] = df['Date'].dt.date
df['Booking_Time'] = df['Date'].dt.time
```

```python
df.head()
```

```python
df.info()
```

```python
# removing unnecessary columns
df.drop(['Date','Time','V_TAT','C_TAT','Canceled_Rides_by_Customer','Canceled_Rides_by_Driver','Incomplete_Rides_Reason'],
         axis=1,inplace=True)
```

```python
# filling null values with mean value of column
df['Customer_Rating'].fillna(df['Customer_Rating'].mean().round())
df['Driver_Ratings'].fillna(df['Customer_Rating'].mean().round())
```

```python
df.info()
```

```python
# converting extracted date colmn to  datetime
df['Booking_Date'] = pd.to_datetime(df['Booking_Date'])
```

```python
# converting a new column from booking date colun
df['Booking_Month'] = df['Booking_Date'].dt.month
```

```python
# making new coulmns
df['Total_Revenue'] = df['Booking_Value'].sum()
```

```python
df['Total_Bookings'] = df['Booking_ID'].count()
```

```python
# save data for further visualiza
df.to_csv('ola_data.csv',index=True,header=True)
```

```python
# 1. Retrieve all successful bookings:
df[df['Booking_Status'] == 'Success']
```

```python
# 2. Find the average ride distance for each vehicle type:
df.groupby('Vehicle_Type')['Ride_Distance'].mean()
```

```python
# 3. Get the total number of cancelled rides by customers:
new_df = df[df['Booking_Status'] == 'Canceled by Customer']
print("Total Number of Cancelled rides:- ",new_df['Booking_Status'].count())
```

```python
# 4. List the top 5 customers who booked the highest number of rides:
df.groupby('Customer_ID')['Booking_ID'].count().sort_values(ascending=False).head(5)
```

```python
# 6. Find the maximum and minimum driver ratings for Prime Sedan bookings:
md = df.groupby('Vehicle_Type')['Driver_Ratings'].max().reset_index()
max_rating = md[md['Vehicle_Type'] == 'Prime Sedan']
mr = df.groupby('Vehicle_Type')['Driver_Ratings'].min().reset_index()
min_rating = mr[mr['Vehicle_Type'] == 'Prime Sedan']
min_rating,max_rating
```

```python
# 7. Retrieve all rides where payment was made using UPI:
df[df['Payment_Method'] == 'UPI']
```

```python
# 8. Find the average customer rating per vehicle type:
df.groupby('Vehicle_Type')['Customer_Rating'].mean()
```

```python
# 9. Calculate the total booking value of rides completed successfully:
df[df['Incomplete_Rides'] == 'No']['Booking_Value'].count()
```
